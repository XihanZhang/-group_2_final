```python
from sklearn import preprocessing
def clean(df):
    df = df.copy()
    
    encode = preprocessing.LabelEncoder()

    for column in df.columns:
        if df[column].dtype == np.object:
            df[column] = df[column].fillna('NaN') #Categorical blank values get set as separate category 
            df.loc[:, column] = encode.fit_transform(df[column])
       
    return df

def nan_fraction(column, df):
    return len(df[df[column].isnull()]) * 1. / len(df)

def impute(df):
    df = df.copy()
    for column in df.columns:
        if nan_fraction(column, df) > 0:
            df[column] = df[column].fillna(df_train_noimp[df_train_noimp[column].notnull()==1][column].mean()) ##using mean to impute  ?
    return df  
```

```python
df_nocat = clean(df)
```

```python
df_nocat.head()
```

```python
np.random.seed(9001)

msk = np.random.rand(len(df_nocat)) < 0.75
df_train_noimp = df_nocat[msk]
df_test_noimp = df_nocat[~msk]
df_test_noimp.shape, df_train_noimp.shape
```

```python
df_train=impute(df_train_noimp)
df_test=impute(df_test_noimp)
df_test.shape, df_train.shape
```

```python
#x_train = df_train.iloc[:,2:-1]
y_train = df_train['DX_bl'].values
#x_test = df_test.iloc[:,2:-1]
y_test = df_test['DX_bl'].values

drop_list = ['RID','DX_bl','PTETHCAT','Ventricles','Hippocampus','WholeBrain','Entorhinal','Fusiform','MidTemp']
x_train = df_train.drop(drop_list,axis = 1).values
x_test = df_test.drop(drop_list,axis = 1).values

#generating quadratic terms
quad_features = preprocessing.PolynomialFeatures(degree = 2)
x_quad = quad_features.fit_transform(df_train.drop(drop_list,axis = 1).iloc[:,3:].values)
x_cat = df_train.drop(drop_list,axis = 1).iloc[:,0:3]
df_x_expanded_train=pd.concat([pd.DataFrame(x_quad, index=x_cat.index), x_cat], axis=1)
x_expanded_train = df_x_expanded_train.values

x_quad_test = quad_features.fit_transform(df_test.drop(drop_list,axis = 1).iloc[:,3:].values)
x_cat_test = df_test.drop(drop_list,axis = 1).iloc[:,0:3]
df_x_expanded_test=pd.concat([pd.DataFrame(x_quad_test, index=x_cat_test.index), x_cat_test], axis=1)
x_expanded_test = df_x_expanded_test.values
```

```python
evaluate_model = lambda model, x_train, y_train: pd.Series([model.score(x_train, y_train), 
                                                 model.score(x_train[y_train==0], y_train[y_train==0]),
                                                 model.score(x_train[y_train==1], y_train[y_train==1]), 
                                                 model.score(x_train[y_train==2], y_train[y_train==2])],
                                                 index=['overall accuracy', 'accuracy on AD', 'accuracy on class CN',
                                                       'accuracy on class LMCI'])
```

```python
#knn tune
from sklearn.model_selection import KFold
n_folds=5
val_score = []
for train_index, val_index in KFold(n_folds, shuffle=True).split(range(len(x_train))):
    X_train, X_val = x_train[train_index], x_train[val_index]
    Y_train, Y_val = y_train[train_index], y_train[val_index]
    
    eachk_score = []
    for k in range(1,12):
        knn_k = KNeighborsClassifier(k)
        knn_k.fit(X_train, Y_train)
        eachk_score.append(knn_k.score(X_val, Y_val))
    val_score.append(eachk_score)

best_k = np.argmax(np.mean(pd.DataFrame(val_score)))+1

knn = KNeighborsClassifier(best_k)
knn.fit(x_train, y_train)
print("k-Nearest Neighbors model: \n")
print("Best k: ", best_k)
```

```python
# tune decision tree
# 5-fold cross validation to pick best maximum depth
from sklearn.model_selection import cross_val_score
score = []
tree_min = 1   
tree_max = 27  
for i in range(tree_min,tree_max):
    dt = DecisionTreeClassifier(max_depth=i)
    # Perform 5-fold cross validation 
    scores = cross_val_score(estimator=dt, X=x_train, y=y_train, cv=5, n_jobs=-1)
    score.append(scores.mean())

# return the index of the best score, depth=index+1
optimal_depth = np.argmax(score)+1 
print("optimal maximum depth: ", optimal_depth)
```

```python
from itertools import product
from collections import OrderedDict
# tune adaboost
param_dict = OrderedDict(
    n_estimators = [30,50,70,100], 
    max_depth = [1,2,10],
    learning_rate = [0.01, 0.05, 0.1]               
) 

all0 = {}
for n, md, r in product(*param_dict.values()):
    params = (n, md, r)
    all0[params] = 0
    dfscore = pd.DataFrame(all0, index=[0])

k=1
for train_index, val_index in KFold(n_folds, shuffle=True).split(range(len(x_train))):
    X_train, X_val = x_train[train_index], x_train[val_index]
    Y_train, Y_val = y_train[train_index], y_train[val_index]
    
    val_score = {}
    for n, md, r in product(*param_dict.values()):
        params = (n, md, r)
        ada=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=md),n_estimators=n, learning_rate=r)
        ada.fit(X_train, Y_train)  
        val_score[params] = ada.score(X_val, Y_val)
        df = pd.DataFrame(val_score, index=[k])
        
    dfscore = dfscore.append(df)
    k=k+1 
```

```python
dfscore.drop([0])
```

```python
md=dfscore.drop([0]).mean().argmax()[1] #max depth
ntree=dfscore.drop([0]).mean().argmax()[0] # number of trees
r=dfscore.drop([0]).mean().argmax()[2] #learning rate
```

```python
fold = 5
C = np.power(10.0, np.arange(-10, 10))

#Unweighted logistic regression
unweighted_logistic = LogisticRegressionCV(Cs=list(C), cv=fold)
unweighted_logistic.fit(x_train, y_train)

unweighted_log_eva = evaluate_model(unweighted_logistic, x_train, y_train)
print('unweighted log')

#Weighted logistic regression
weighted_logistic = LogisticRegressionCV(Cs=list(C), cv=fold, class_weight='balanced')
weighted_logistic.fit(x_train, y_train)

weighted_log_eva = evaluate_model(weighted_logistic, x_train, y_train)
print('weighted log')


#Unweighted Logistic Regression with quadratic terms
unweighted_logreg_poly = LogisticRegressionCV(Cs=list(C), cv=fold)
unweighted_logreg_poly.fit(x_expanded_train, y_train)
    
unweighted_log_poly_eva = evaluate_model(unweighted_logreg_poly, x_expanded_train, y_train)
print('unweighted log with quadratic terms')

#Weighted Logistic Regression with quadratic terms
weighted_logreg_poly = LogisticRegressionCV(Cs=list(C), cv=fold, class_weight='balanced')
weighted_logreg_poly.fit(x_expanded_train, y_train)
    
weighted_log_poly_eva = evaluate_model(weighted_logreg_poly, x_expanded_train, y_train)
print('weighted log with quadratic terms')

#LDA
lda = LDA()
lda.fit(x_train, y_train)

lda_eva = evaluate_model(lda, x_train, y_train)
print('lda')

#QDA
qda = QDA()
qda.fit(x_train, y_train)

qda_eva = evaluate_model(qda, x_train, y_train)
print('qda')

#Decision Tree
tree = DecisionTree(max_depth=optimal_depth, class_weight='balanced', criterion='entropy')
tree.fit(x_train, y_train)

tree_eva = evaluate_model(tree, x_train, y_train)
print('decision tree')

#KNN
knn = KNN(n_neighbors=best_k) 
knn.fit(x_train, y_train)

knn_eva = evaluate_model(knn, x_train, y_train)
print('knn')

#random forest
param_dict = OrderedDict(
    max_depth = [1,2,5,10],
    n_estimators = [30,50,70,100],
    max_features = [0.2, 0.4, 0.6, 0.8])
from sklearn.model_selection import GridSearchCV
est = RandomForestClassifier(oob_score=False)
gs = GridSearchCV(est, param_grid = param_dict, cv=5, n_jobs=-1)
gs.fit(x_train, y_train)
rf = gs.best_estimator_

rf_eva = evaluate_model(rf, x_train, y_train)
print('random forest')

#adaboost
ada=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=md), n_estimators=ntree, learning_rate=r)
ada.fit(x_train, y_train)

ada_eva = evaluate_model(ada, x_train, y_train)
print('adaboost')

#linear SVM
#cvals = [1e-20, 1e-15, 1e-10, 1e-5, 1e-3, 1e-1, 1, 10, 100, 10000, 100000] #?
#parameters = {'C': cvals}
#est = svm.SVC(kernel='linear')
#gs = GridSearchCV(est, param_grid=parameters, cv=5) #n_job?
#gs.fit(x_train, y_train)
#svm_linear = gs.best_estimator_
#svm_linear_eva = evaluate_model(svm_linear, x_train, y_train)
#print("linear SVM")

#linear SVM
#svm_linear = svm.SVC(C=1000, kernel='linear')
#svm_linear.fit(x_train, y_train)

#svm_linear_eva = evaluate_model(svm_linear, x_train, y_train)
#print("linear SVM")

#quadratic SVM
#svm_poly = svm.SVC(C=1000, kernel='poly', degree=2)
#svm_poly.fit(x_train, y_train)

#svm_poly_eva = evaluate_model(svm_poly, x_train, y_train)
#print("quadratic SVM")
```
